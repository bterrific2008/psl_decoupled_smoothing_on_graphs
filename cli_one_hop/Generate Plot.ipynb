```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
```

```python
import re
labeled = []
auc = []
with open('auc.txt', 'r') as f:
    for line in f:
        u, a = (re.split(r' +', line.strip()))
        labeled.append(1-float(u))
        auc.append(float(a))
labeled, auc
```

```python
plt.plot(labeled, auc, color='r', label='1-hop MV')
plt.xlabel('Percentage of Nodes Initially Labeled')
plt.ylabel('AUC')
plt.title('Amherst')
plt.legend()
plt.savefig('1_hop.jpg')
```
